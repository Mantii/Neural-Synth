```python
import torchaudio
import numpy as np
from os import listdir
from IPython.display import Audio
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d='audio/guitar/'
for i in listdir(d):
    d+=i
    break
sample=torchaudio.load(d)[0].view(64000).numpy()
print(len(sample))
```

```python
Audio(sample,rate=16000)
```

```python
plt.plot(sample)
```

```python
stride=2
kernel=np.array([1,0,5,0,0,0,1])
y=np.zeros(len(sample))
for i in range(0,len(sample)-len(kernel),stride):
    bit=sample[i:i+len(kernel)]
    c=np.sum(bit*kernel)
    y[i]=c
    
```

```python
section=500
plt.plot(sample[:section],'g')
plt.plot(y[:section],'r')
plt.show()
```
