```python
import torch
from torch.autograd import Variable
import numpy as np
import torch.nn.functional as F
import torch.nn as nn
import torchvision
from torchvision import transforms
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import tqdm
from downModel import AutoEncoder
from utils import load_dataset
import IPython
from IPython.display import Audio
```

```python
from torch.utils.data import DataLoader
dataloader=DataLoader(load_dataset(samples=400),batch_size=1240,shuffle=False)
```

```python
vae = AutoEncoder().cuda()


checkpoint = torch.load('../wavenet/batchMSE.tar')
vae.load_state_dict(checkpoint['vae'])
print("\nLoaded model checkpoint\nEpoch: ",checkpoint['epoch'],'\nLoss:',checkpoint['loss'])
```

```python
labels=[]
vecs=[]
for data in tqdm(dataloader):
    x=Variable(data.cuda())
    labels.append(x[:,:,1024].cpu().detach().numpy())
    vae.zero_grad()
    y,z=vae(x)
    vecs.append(z.cpu().detach().numpy().reshape(20,128*62))
```

```python
zs=[]
label=[]
for batchz in vecs:
    for i in batchz:
        print(i.shape)
        zs.append(i)
        label.append(int(len(zs)/400))
    
    
```

```python
b=0
for i,j in zip(zs,label):
    if b%10==0:
        print(b,i.shape,j)
    b+=1
```
